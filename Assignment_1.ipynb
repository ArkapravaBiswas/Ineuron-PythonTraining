```python
""" 2.
Write a program which will find all such numbers which are divisible by 7 but are not a multiple
of 5, between 2000 and 3200 (both included). The numbers obtained should be printed in a
comma-separated sequence on a single line."""


result_list=[]
for number in range(2000,3201):
    if ((number%7==0) and (number%5!=0)):
        result_list.append(number)
for final_number in result_list:
    print(final_number,end=',')
```

```python
"""3.
Write a Python program to accept the user's first and last name and then getting them printed in
the the reverse order with a space between first name and last name."""

input_list=[]
output_list=[]
formatted_output=''
input_list.append(input('enter the first name: '))
input_list.append(input('enter the second number: '))
input_list.reverse()
for i in input_list:
    output_list.append(i[::-1])
print(output_list)
for i in output_list:
    formatted_output=formatted_output+i+" "
print("the required output: ",formatted_output)

```

```python
"""2.
Create the below pattern using nested for loop in Python.
*
* *
* * *
* * * *
* * * * *
* * * *
* * *
* *
*"""

number_of_rows=5;
for i in range(number_of_rows):
    for j in range(i):
        print ('* ', end="")
    print('')

for i in range(number_of_rows,0,-1):
    for j in range(i):
        print('* ', end="")
    print('')
```
